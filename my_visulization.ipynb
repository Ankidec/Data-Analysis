Matplotlib Vs Seaborn Vs Plotly

```python
import matplotlib.pyplot as plt
x=[1,2,3,5,6,7,8,10]
y=[7,2,1,5,6,9,10,23]
plt.plot(x, y, color='r')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
import seaborn as sns
fig = sns.lineplot(x, y, color='r')
fig.set(xlabel='x-axis',ylabel='y-axis')
fig.grid()
fig.set_title('Line plot')
```
