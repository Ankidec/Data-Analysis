```python
import matplotlib.pyplot as plt # now the pyplot package can be reffered as plt.
import numpy as np
#plotting x and y 
#plot() fuction is used to drow point or markers in a diagram
 
xpoints = np.array([0,6]) # first bydefalt x asix hota h 
ypoints= np.array([0,100])# second y asix
 

plt.plot(xpoints,ypoints)
plt.show() 
```

```python
# plotting with out line 
xpoints = np.array([0,6])  
ypoints= np.array([0,10])

plt.plot(xpoints,ypoints, 'o')
plt.show() 
```

```python
#multipal value
xpoints = np.array([1,2,6,8]) 
ypoints= np.array([3,8,1,10])
 

plt.plot(xpoints,ypoints)
plt.show() 
```

```python
#Defalt X- point
# if we do not specify the point in x asix then the defalt values 0,1,2,3,4....
ypoint = np.array([3,1,4,8,10])
plt.plot(ypoint)
plt.show
```

```python
#markers( we can used argument to emphasize each point with spesified marker)
import matplotlib.pyplot as plt
import numpy as np
ypoint = np.array([3,8,1,10])
plt.plot(ypoint, marker= 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['python','C','C++','Java','C#','Html']
y=[85,70,60,82,65,83]
plt.xlabel('Language',fontsize=10)
plt.ylabel('No of Popularity',fontsize=10)
plt.title('Language of Programming',fontsize=25)
c='b','r','y','g','m','b'
plt.bar(x,y,color=c,width=0.4,align='edge',edgecolor='r')#edge color se baunder ki color change ho jata h
plt.show
```

```python
import matplotlib.pyplot as plt
x=['python','C','C++','Java','C#','Html']
y=[85,70,60,82,65,83]
plt.xlabel('Language',fontsize=10)
plt.ylabel('No of Popularity',fontsize=10)
plt.title('Language of Programming',fontsize=25)
c='b','r','y','g','m','b'
plt.bar(x,y,color=c,width=0.4,edgecolor='r',alpha=0.5,label='Popukarity')
plt.legend()
plt.show
```

Scatter Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[9,8,7,6,5,4,3]
plt.scatter(x,y)
plt.show
```

```python
day=[1,2,3,4,5,6,7]
no=[9,8,3,1,5,4,3]
colors=['r','y','g','b','r','g','r']
sizes=[50,60,70,30,20,56,78]
plt.scatter(day,no,c=colors,s=sizes)
plt.title("Scatter plot",fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.show
```
