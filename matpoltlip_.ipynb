```python
import matplotlib.pyplot as plt # now the pyplot package can be reffered as plt.
import numpy as np
#plotting x and y 
#plot() fuction is used to drow point or markers in a diagram
 
xpoints = np.array([0,6]) # first bydefalt x asix hota h 
ypoints= np.array([0,100])# second y asix
 

plt.plot(xpoints,ypoints)
plt.show() 
```

```python
# plotting with out line 
xpoints = np.array([0,6])  
ypoints= np.array([0,10])

plt.plot(xpoints,ypoints, 'o')
plt.show() 
```

```python
#multipal value
xpoints = np.array([1,2,6,8]) 
ypoints= np.array([3,8,1,10])
 

plt.plot(xpoints,ypoints)
plt.show() 
```

```python
#Defalt X- point
# if we do not specify the point in x asix then the defalt values 0,1,2,3,4....
ypoint = np.array([3,1,4,8,10])
plt.plot(ypoint)
plt.show
```

```python
#markers( we can used argument to emphasize each point with spesified marker)
import matplotlib.pyplot as plt
import numpy as np
ypoint = np.array([3,8,1,10])
plt.plot(ypoint, marker= 'o')
plt.show()
```
