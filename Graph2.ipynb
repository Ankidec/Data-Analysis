# categorical graphs
- columns that have text are generally categorical
- the first step is to count the number of unique 
- then plot the count of each category
- following chart r a good way to visualize categorical data
     - bar chart 
     - pie chart 
     - dount chart 
     - line chart(when too many categories)
     - area chart(when too many categories) 

```python
titanic['survived'].value_counts()
```

```python
survival_count = titanic['survived'].value_counts()
survival_count.plot(kind = 'pie')
```

```python
survial_count = titanic['alive'].value_counts()
survial_count.plot(kind='pie', autopct='%.1f%%', title='survial rate')
```

```python
survival_count = titanic['survived'].value_counts()
survival_count.plot(kind = 'pie', autopct='%.1f%%')
```

```python
p_class = titanic['pclass'].value_counts()
p_class.plot(kind = 'pie', autopct='%.1f%%')
```

```python
survial_count.plot(
    kind='pie',
    autopct='%.1f%%',
    title='survival rate',
    labels=['Survived', 'Not Survived'],
    colors=['#66b3ff','#ff9999'],
    labeldistance=1.1,
    wedgeprops={
        'width': 0.5,
    }
)
```

```python
p_class = titanic['pclass'].value_counts()
p_class
```

```python
p_class.plot(
    kind = 'pie',
    autopct = '%.1f%%',
    title = 'Number of peolpe in different levels of ships',
    labels = ['3','1','2'],
    colors = ['#66b3ff','#ff9999','#ffbb11'],
    labeldistance=1.1,
    wedgeprops={
        'width' : 0.5,
    })
```
