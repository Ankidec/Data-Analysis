```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('drinking_water_potability.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# histogram
sns.histplot(data=df, x="ph", bins=[0,2,4,6,8,10,12,14])
```

```python
user_inp = input(f'select a column to generate histogram from the list {df.columns.tolist()}: ')
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15,5))    
if user_inp in df.columns.tolist():
    sns.histplot(data=df, x=user_inp, ax=ax1)
    sns.boxplot(data=df, x=user_inp, ax=ax2)
plt.show()
```

```python
sns.relplot(data=df, x='Hardness', y='ph', hue='Potability')
plt.savefig('hardness_ph.png', dpi=300, bbox_inches='tight')
```

```python
import plotly.express as px
px.scatter_3d(df, x='Hardness', y='ph', z='Solids', color='Potability')
```
